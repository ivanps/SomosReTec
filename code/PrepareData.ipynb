# Sustainable Behavior Questionnaire Data

This notebook cleans and processes the sustainable behavior questionnaire data. The dataset includes four sections:
- **Proecological behavior** (16 questions)
- **Frugality behavior** (10 questions)
- **Equity** (7 questions)
- **Altruism** (10 questions)

The cleaning process involves:
1. Loading the raw data
2. Renaming columns to standardized format
3. Creating group variables (Control vs Experimental)
4. Converting Likert scale text responses to numeric values

## Load libraries and data

```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

Load the raw Excel data and display the column names for each questionnaire section to verify the structure.

Configure pandas display settings to show all rows and columns when viewing dataframes.

```python
df0 = pd.read_excel('..\data\Conductas_Sustentables_Dataset.xlsx')
print(df0.columns[5:21])
print(df0.columns[21:31])
print(df0.columns[31:38])
print(df0.columns[38:48])
```

Prepare the data structure:
- Rename all questionnaire columns to standardized format (proeco1-16, frugal1-10, equity1-7, altruism1-10)

```python
# Rename columns to prepare analysis
cols = list(df0.columns)
cols[5:21] = [f'proeco{i}' for i in range(1, 17)]
cols[21:31] = [f'frugal{i}' for i in range(1, 11)]
cols[31:38] = [f'equity{i}' for i in range(1, 8)]
cols[38:48] = [f'altruism{i}' for i in range(1, 11)]
df0.columns = cols
df0.head()
```

## Clean data groups

```python
from datetime import datetime

# Make a copy of the original dataframe to work with
df = df0.copy()

df['Timestamp'] = df['Timestamp'].astype(str).apply(
    lambda x: datetime.strptime(x[:-4], '%Y/%m/%d %I:%M:%S %p')
)
df['month'] = df['Timestamp'].apply(lambda x: x.strftime('%B'))
df['month'].value_counts()
```

```python
df['Momento'] = df['Momento'].str.strip().str.capitalize()
df['Momento'] = df['Momento'].replace({'Pps': 'Post'})
df['Momento'].value_counts()
```

```python
df['Tipo de grupo'] = df['Tipo de grupo'].str.strip().str.capitalize()
df['Tipo de grupo'].value_counts()
```

```python
df['Profesor del grupo'].value_counts()
```

```python
# Cross tabulation of 'Momento' and 'month'
pd.crosstab(df['month'], df['Momento'])
```

```python
# Print the Momeento=Pre from month=October and the Post from November
iflag = (df['Momento'] == 'Pre') & (df['month'] == 'October') | (df['Momento'] == 'Post') & (df['month'] == 'November')
df.loc[iflag, df.columns[:5]]
```

## Code Likert questions

Convert all text-based Likert scale responses to numeric values for statistical analysis.

```python
# First eliminate blank spaces in all likert responses (columns 5 to 47 inclusive)
df.loc[:, df.columns[5:48]] = df.loc[:, df.columns[5:48]].map(lambda x: x.strip() if isinstance(x, str) else x)
```

### Proecological behavior questions
df.iloc[:, 5:48] = df.iloc[:, 5:48].applymap(lambda x: x.strip() if isinstance(x, str) else x)
Code proecological questions (proeco1-proeco16) to numeric scale (0-3):
- **0** = Nunca (Never)
- **1** = Casi Nunca (Almost Never)
- **2** = Casi Siempre (Almost Always)
- **3** = Siempre (Always)

**Note:** proeco1 and proeco2 include special handling for text responses like "No la lavo" (I don't wash it), which are coded as 0.

```python
for i in range(1, 3):
    col_name = f'proeco{i}'
    df[col_name] = df[col_name].replace({
        'No lavo': 0,
        'Mi mamÃ¡ lava la ropa, pero creo que si': 0,
        'No lavo, pero me imagino que asi lo haria': 0,
        'no lavo yo mi ropa': 0,
        'No la lavo': 0,
        'No lavo mi ropa': 0,
        'Yo no lavo la ropa': 0,
        'Yo no lavo': 0,
        '0= Nunca': 0,
        '1= Casi Nunca': 1,
        '2= Casi Siempre': 2,
        '3= Siempre': 3
    })
for i in range(2, 17):
    col_name = f'proeco{i}'
    df[col_name] = df[col_name].replace({
        '0= Nunca': 0,
        '1= Casi Nunca': 1,
        '2= Casi Siempre': 2,
        '3= Siempre': 3
    })
df.loc[:, 'proeco1':'proeco16'].head()
```

### Frugality behavior questions

Code frugality questions (frugal1-frugal10) to numeric scale (0-4):
- **0** = Nunca (Never)
- **1** = Casi Nunca (Almost Never)
- **2** = A veces (Sometimes)
- **3** = Casi Siempre (Almost Always)
- **4** = Siempre (Always)

```python
for i in range(1, 11):
    col_name = f'frugal{i}'
    df[col_name] = df[col_name].replace({
        '0= Nunca': 0,
        '1= Casi Nunca': 1,
        '2= A veces' : 2,
        '3= Casi Siempre': 3,
        '4= Siempre': 4
    })
df.loc[:, 'frugal1':'frugal10'].head()
```

### Equity questions

Code equity questions (equity1-equity7) to numeric agreement scale (1-5):
- **1** = Totalmente en desacuerdo (Totally Disagree)
- **2** = Parcialmente en desacuerdo (Partially Disagree)
- **3** = No estoy seguro (Not Sure)
- **4** = Parcialmente de acuerdo (Partially Agree)
- **5** = Totalmente de acuerdo (Totally Agree)

**Note:** equity1 has slightly different text format than equity2-7, so it's coded separately.

```python
df['equity1'] = df['equity1'].replace({
    '1=Totalmente en desacuerdo': 1,
    '2=Parcialmente en desacuerdo': 2,
    '3=No estoy seguro' : 3,
    '4=Parcialmente de acuerdo': 4,
    '5=Totalmente de acuerdo': 5
})

for i in range(2, 8):
    col_name = f'equity{i}'
    df[col_name] = df[col_name].replace({
        '1= Totalmente en desacuerdo': 1,
        '2= Parcialmente en desacuerdo': 2,
        '3= No estoy seguro' : 3,
        '4= Parcialmente de acuerdo': 4,
        '5= Totalmente de acuerdo': 5
    })
df.loc[:,'equity1':'equity7'].head()
```

### Altruism questions

Code altruism questions (altruism1-altruism10) to numeric scale (1-4):
- **1** = Nunca (Never)
- **2** = Casi Nunca (Almost Never)
- **3** = Casi Siempre (Almost Always)
- **4** = Siempre (Always)

```python
for i in range(1, 11):
    col_name = f'altruism{i}'
    df[col_name] = df[col_name].replace({
        '1= Nunca': 1,
        '2= Casi Nunca': 2,
        '3= Casi Siempre' : 3,
        '4= Siempre': 4
    })
df.loc[:, 'altruism1':'altruism10'].head()
```

## Sustainable Behavior Instruments

```python
# Compute proeco score as the average of proeco1 to proeco16
df['proecologica'] = df.loc[:, 'proeco2':'proeco16'].mean(axis=1)
# Compute frugal score as the average of frugal1 to frugal10
df['frugalidad'] = df.loc[:, 'frugal1':'frugal10'].mean(axis=1)
# Compute equity score as the average of equity1 to equity7
df['equidad'] = df.loc[:, 'equity1':'equity7'].mean(axis=1)
# Compute altruism score as the average of altruism1 to altruism10
df['altruismo'] = df.loc[:, 'altruism1':'altruism10'].mean(axis=1)
df[['proecologica', 'frugalidad', 'equidad', 'altruismo']].head()
```
